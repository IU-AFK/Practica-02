```python
playlist = [
{"title": "Bohemian Rhapsody", "duration": "5:55"},
{"title": "Hotel California", "duration": "6:30"},
{"title": "Stairway to Heaven", "duration": "8:02"},
{"title": "Imagine", "duration": "3:07"},
{"title": "Smells Like Teen Spirit", "duration": "5:01"},
{"title": "Billie Jean", "duration": "4:54"},
{"title": "Hey Jude", "duration": "7:11"},
{"title": "Like a Rolling Stone", "duration": "6:13"},
]

all_songs = {}
for song in playlist:
    minutes,seconds = song["duration"].split(":")
    ts_song = int(minutes)*60 + int(seconds)
    all_songs[song["title"]] = ts_song

total_duration = sum(all_songs.values())
max_song = max(all_songs, key=all_songs.get)
min_song = min(all_songs, key=all_songs.get)

print(f"Duracion total de la playlist: {int(total_duration/60)}m {total_duration%60}s")
print(f"Cancion mas larga: {max_song} ({int(all_songs[max_song]/60)}:{all_songs[max_song]%60:02d})")
print(f"Cancion mas corta: {min_song} ({int(all_songs[min_song]/60)}:{all_songs[min_song]%60:02d})")
```
