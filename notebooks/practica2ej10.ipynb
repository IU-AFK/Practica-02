```python
p_stats = {
    'Valentina': {'wins':0,'max':0,'prom':0,'score':0},
    'Mateo': {'wins':0,'max':0,'prom':0,'score':0},
    'Camila': {'wins':0,'max':0,'prom':0,'score':0},
    'Santiago': {'wins':0,'max':0,'prom':0,'score':0},
    'Lucía': {'wins':0,'max':0,'prom':0,'score':0},
}

for i,r in enumerate(rounds):
    r_winner ={
    'player':' ',
    'score': -1
}
    for player in r['scores']:
        round_scr = sum(r['scores'][player].values())  #puntaje total de la ronda
        p_stats[player]['score'] += round_scr  #acumulacion de puntaje total
        
        if p_stats[player]['max'] < round_scr:  #mejor puntaje en una ronda
            p_stats[player]['max'] = round_scr
        if r_winner['score'] < round_scr:  #ganador de la ronda
            r_winner['score'] = round_scr
            r_winner['player'] = player
        p_stats[player]['prom'] = p_stats[player]['score']/len(rounds)  #calcula promedios
        
    p_stats[r_winner['player']]['wins'] += 1

    print(f"Ronda {i+1} - {r['theme']}: \n Ganador: {r_winner['player']} *{r_winner['score']}*")
    if r['theme'] != 'Final libre':
        print_format_table(p_stats,"Tabla de Posiciones")
    else:
        print_format_table(p_stats,"Tabla de Posiciones Final")
```
